```python
# Dependencies
import pandas as pd
import re
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Major_League_Soccer_stadiums#Stadiums'
```

```python
tables = pd.read_html(url)
tables
```

```python
df1 = tables[1]
df2 = tables[7]
df3 = tables[3]
for col in df1.columns:
    print(col)
```

```python
clean_df1 = df1.drop(columns=['Image','Opened','Field dimensions','Coordinates','Soccer specific','Ref(s)','First MLS year in stadium'])
clean_df2 = df2.drop(columns=['Image','Opened','Ref(s)','Used for MLS'])
clean_df3 = df3.drop(columns=['Image','Opened','Soccer specific','Ref(s)','Year Since Use'])
clean_df2['Roof type'] = 'Open'
```

```python
clean_df1.head()
```

```python
clean_df2.head()
```

```python
clean_df3.head()
```

```python
clean_df1 = clean_df1.rename(columns={"Franchise": "Franchise(s)","Roof type":"Roof Type"})
clean_df2 = clean_df2.rename(columns={"Franchise": "Franchise(s)","Roof type":"Roof Type"})
clean_df3 = clean_df3.rename(columns={"Franchise": "Franchise(s)"})
```

```python
final_df = pd.concat([clean_df1,clean_df2,clean_df3])
final_df
```

```python
new = final_df["Location"].str.split(",", n = 1, expand = True) 
final_df["City"]= new[0] 
final_df["State"]= new[1]
final_df.drop(columns =["Location"], inplace = True) 
final_df.reset_index(drop=True)
```

```python
final_df["Capacity"]=final_df["Capacity"].str.replace(r"\[.*\]","")
final_df
```

```python
final_df=final_df.reset_index(drop=True)
final_df
```

```python
for i,r in final_df.iterrows():
    value = final_df.loc[i, "Capacity"]
    cap = re.findall('\(.*?\)',value)
    if cap == []:
        continue
    else:
        final_df.loc[i, "Capacity"] = cap[0]
final_df['Capacity'] = final_df['Capacity'].str.replace(r'\D', '')
final_df
```

```python
final_df.to_csv('stadium_data.csv')
```
