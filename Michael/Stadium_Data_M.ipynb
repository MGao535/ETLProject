```python
# Dependencies
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Major_League_Soccer_stadiums#Stadiums'
```

```python
tables = pd.read_html(url)
tables
```

```python
tables[1]
```

```python
tables[7]
```

```python
df1 = tables[1]
df2 = tables[7]
for col in df1.columns:
    print(col)
```

```python
clean_df1 = df1.drop(columns=['Image','Opened','Field dimensions','Coordinates','Soccer specific','Ref(s)'])
clean_df2 = df2.drop(columns=['Image','Opened','Ref(s)'])
clean_df2['Roof type'] = 'Open'
```

```python
clean_df1.head()
```

```python
clean_df2.head()
```

```python
clean_df1 = clean_df1.rename(columns={"Franchise": "Franchise(s)", "First MLS year in stadium": "Used for MLS"})
```

```python
final_df = pd.concat([clean_df1,clean_df2])
final_df
```

```python
new = final_df["Location"].str.split(",", n = 1, expand = True) 
final_df["City"]= new[0] 
final_df["State"]= new[1]
final_df.drop(columns =["Location"], inplace = True) 
final_df.reset_index(drop=True)
```

```python
final_df["Capacity"]=final_df["Capacity"].str.replace(r"\[.*\]","")
final_df
```
