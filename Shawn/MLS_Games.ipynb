```python
import pandas as pd 
from datetime import datetime, date, time, timezone
pd.options.display.float_format = '{:.2f}'.format
```

```python
file = 'MLSAttendanceAllGames.csv'
table = pd.read_csv(file)
```

```python
table
```

```python
table.dropna(inplace=True)
table 
```

```python
mls_game = table[['Game #', 'Date', 'Home', 'Home Parent', 'Home G', 'Visitor Parent', 'Visitor G', 'Att', 'Staduim', 'Percentage Filled']].copy()
```

```python
mls_game
```

```python
mls_game.columns = ['Game #', 'Date', 'City', 'Home Team', 'Home Goals', 'Visitor Team', 'Visitor Goals', 'Attendance', 'Stadium', 'Percentage Filled']
mls_game
```

```python
dt = mls_game['Date'].apply(lambda x: datetime.strptime(x,'%m/%d/%y'))
dt
```

```python
dt1 = pd.to_datetime(dt).astype(int) / 10**9
```

```python
mls_game = mls_game.assign(Unix = dt1)
mls_game
```

```python
mls_game.to_csv('mls_game.csv')
```

```python
mls_2013_2018 = mls_game.loc[mls_game['Unix'] > 1362182400]
```

```python
mls_2013_2018
```

```python
mls_2013_2018.to_csv('mls_2013_2018.csv')
```
